# Modul 2: Operator Matematika & Manipulasi String

Setelah memahami variabel, sekarang kita akan belajar bagaimana mengolah data tersebut.
Fokus kita kali ini adalah:
1. Menggunakan Python sebagai kalkulator (Operator).
2. Menggabungkan dan memodifikasi teks (String).

```python
# 1. Operator Aritmatika
a = 10
b = 3

print("--- Hasil Perhitungan ---")
print("Penjumlahan (10 + 3) :", a + b)
print("Pengurangan (10 - 3) :", a - b)
print("Perkalian (10 * 3) :", a * b)
print("Pembagian (10 / 3) :", a / b)
print("Pembagian Bulat (10 // 3) :", a // b) # Menghilangkan angka di belakang koma
print("Sisa Bagi / Modulo (10 % 3) :", a % b) # Berguna untuk cek angka ganjil/genap
print("Perpangkatan (10 ^ 3) :", a ** b)
```

## 2. Manipulasi String
String bukan sekadar teks. Di Python, kita bisa melakukan banyak hal menarik dengan string, seperti menggabungkannya atau mengubah formatnya.

```python
# Menggabungkan String (Concatenation)
depan = "Python"
belakang = "Expert"
full_name = depan + " " + belakang
print("Gabungan:", full_name)

# Perkalian String (repetition)
print("Semangat! " * 3)

# String Methods (Mengubah Format)
pesan = "belajar python itu seru"
print(pesan.upper())   # jadi huruf besar semua
print(pesan.title())   # jadi huruf besar di setiap kata
print(len(pesan))      # menghitung jumlah karakter (termasuk spasi)
```

## 3. F-String (Cara Modern Cetak Variabel)
F-String adalah cara paling direkomendasikan untuk memasukkan variabel ke dalam teks agar lebih rapi.

```python
nama = "Ahmad"
progres = 2
target = 10

# Menggunakan f-string (tambahkan huruf 'f' sebelum tanda kutip)
print(f"Halo {nama}, kamu sudah menyelesaikan {progres} modul dari total {target} modul.")
```
