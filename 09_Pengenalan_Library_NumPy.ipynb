# Modul 9: Pengenalan Library NumPy

NumPy adalah library dasar untuk komputasi sains di Python.
Struktur data utama di NumPy disebut **ndarray (N-dimensional array)**.

Kelebihan NumPy Array dibanding List Python biasa:
1. **Lebih Cepat**: Proses komputasi dilakukan di level memori yang lebih rendah (menggunakan bahasa C).
2. **Vectorization**: Kita bisa melakukan operasi matematika langsung ke semua elemen array tanpa perlu menggunakan perulangan (`for loop`).

```python
# 1. IMPORT LIBRARY NUMPY
# 'as np' adalah alias standar di indistri agar kita tidak perlu mengetik 'numpy' secara panjang
import numpy as np

# Membuat NumPy Array dari List biasa
list_biasa = [10, 20, 30, 40]
arr = np.array(list_biasa)

print("--- Dasar NumPy Array ---")
print("Isi Array     :", arr)
print("Tipe Objek    :", type(arr))
print("Tipe Data Isi :", arr.dtype) # Mengetahui tipe data di dalam array (misal: int64)
```

## 2. Operasi Matematika Otomatis (Vectorization)
Hebatnya NumPy, kita bisa mengalikan atau menambah seluruh isi array secara instan tanpa bantuan `for loop`.

```python
# Perbandingan kemudahan operasi matematika
arr_harga = np.array([10000, 20000, 50000])

# Semua elemen otomatis di kali 2!
arr_dikali_dua = arr_harga * 2

# Semua elemen otomatis di tambah 5000!
arr_ditambah_bonus = arr_harga + 5000

print("--- Operasi Vektor NumPy ---")
print("Harga Awal   :", arr_harga)
print("Dikali dua   :", arr_dikali_dua)
print("Ditambah 5K  :", arr_ditambah_bonus)
```

## 3. Fungsi Statistik Bawaan NumPy
NumPy menyediakan fungsi-fungsi siap pakai untuk menganalisis data angka dengan sangat cepat.

```python
nilai_mahasiswa = np.array([80, 75, 90, 85, 60, 95])

print("--- Analisis Statistik ---")
print("Nilai Tertinggi (Max)   :", np.max(nilai_mahasiswa))
print("Nilai Terendah (Min)    :", np.min(nilai_mahasiswa))
print("Rata-rata (Mean)        :", np.mean(nilai_mahasiswa))
print("Total Jumlah (Sum)      :", np.sum(nilai_mahasiswa))
```

## 📝 Tantangan Modul 9 (Analisis Data Penjualan)
Bayangkan Anda sedang memegang data total penjualan harian dari sebuah toko selama 5 hari:
1. Buatlah sebuah NumPy Array bernama `penjualan` yang berisi data angka: `[150, 300, 250, 400, 100]` (dalam ribuan Rupiah).
2. Gunakan fungsi statistik bawaan NumPy untuk mencari:
   - Rata-rata penjualan harian (simpan di variabel `rata_rata`).
   - Total penjualan selama 5 hari (simpan di variabel `total_penjualan`).
3. Cetak hasilnya menggunakan f-string:
   *"Rata-rata penjualan toko: {rata_rata} ribu per hari, dengan total omset: {total_penjualan} ribu."*

```python
# Jawaban modul 9
penjualan = np.array([150, 300, 250, 400, 100])
rata_rata = np.mean(penjualan)
total_penjualan = np.sum(penjualan)

print("--- Data Penjualan ---")
print(f"Rata-rata penjualan toko: {rata_rata} ribu per hari, dengan total omset: {total_penjualan} ribu.")
```
