# Modul 8: Penanganan Error (Exception Handling)

Di Python, error yang terjadi saat program berjalan disebut **Exception**. Jika tidak ditangani, program akan langsung berhenti terputus (*crash*).

Untuk mencegah hal itu, kita menggunakan blok:
1. `try`: Blok kode yang ingin kita uji (yang berpotensi error).
2. `except`: Blok kode yang akan dieksekusi *hanya jika* terjadi error di blok try.
3. `finally`: Blok kode yang akan *selalu* dijalankan, baik terjadi error maupun tidak (biasanya untuk menutup koneksi database atau file).

```python
# 1. SIMULASI ERROR TANPA JARING PENGAMAN
# jika anda menjalankan kode: print(10 / 0), Python akan crash dengan error "ZeroDivisionError"

# SEKARANG KITA PASANG JARING PENGAMAN (Try-Except)
try:
  angka = 10
  pembagi = 0
  hasil = angka / pembagi
  print(f"Hasil pembagian: {hasil}")
except ZeroDivisionError:
  # Kode ini menyelamatkan program dari crash akibat pembagian dengan 0
  print("Error: Hei, dalam matematika anda tidak boleh membagi angka dengan nol!")

print("Sistem aman, program tetap berlanjut ke baris ini...")
```

## 2. Menangani Berbagai Jenis Error (Multi-Except)
Input dari user sering kali tidak tertebak. Kita bisa memasang beberapa skenario `except` sekaligus untuk jenis error yang berbeda.

```python
# 2. MENANGANI INPUT YANG SALAH TIPE (ValueError)
input_user = "sepuluh"

try:
  # Mengubah string menjadi integer pasti error jika isinya teks
  angka_konversi = int(input_user)
  print(f"Angka Anda adalah: {angka_konversi}")
except ValueError:
  print("Error: Input yang anda masukkan bukan angka yang valid!")
except Exception as e:
  # Pancingan terakhir untuk menangkap semua jenis error lainnya yang tidak terduga
  print(f"Terjadi error umum: {e}")
```

## 3. Menggunakan Blok `finally`
Blok `finally` sangat berguna untuk proses pembersihan (cleanup) karena kodenya dijamin akan selalu dieksekusi oleh Python dalam kondisi apa pun.

```python
try:
  print("Membuka koneksi data...")
  # Jalankan kode acak yang aman
  skor = 100 / 2
except ZeroDivisionError:
  print("Terjadi Error!")
finally:
  # Bagian ini akan selalu muncul di layar
  print("Selesai! Koneksi data berhasil ditutup dengan aman.")
```

## 📝 Tantangan Modul 8 (Sistem Validasi Umur)
Bayangkan Anda sedang membuat form pendaftaran online yang mewajibkan input umur dalam bentuk angka bulat (integer):
1. Buatlah variabel `input_umur = "dua puluh tiga"`. (Sengaja diisi string teks agar memicu error).
2. Buatlah blok **`try-except`**:
   - Di dalam blok `try`, coba ubah variabel `input_umur` tersebut menjadi integer menggunakan fungsi `int()` dan simpan ke variabel baru `umur_valid`.
   - Jika proses konversi berhasil, cetak kalimat dengan f-string: *"Pendaftaran berhasil! Umur Anda terverifikasi: {umur_valid} tahun."*
   - Di dalam blok `except ValueError`, tangkap error-nya dan cetak kalimat penolakan: *"Pendaftaran gagal! Silakan masukkan umur menggunakan angka saja (contoh: 23)."*

```python
# Jawaban modul 8
input_umur = "dua puluh tiga" # sengaja menuliskan dengan bentuk string supaya terjadi error

try:
  umur_valid = int(input_umur)
  # pada kode di bawah pastikan untuk menggunakan f-string karena terdapat variabel yang akan kita panggil
  print(f"Pendaftaran berhasil! Umur Anda terverifikasi: {umur_valid} tahun.")
except ValueError:
  # boleh menggunakan f-string atau tidak, seharusnya tidak perlu karena tidak adanya variabel yg akan di panggil
  print(f"Pendaftaran gagal! Silakan masukkan umur menggunakan angka saja (contoh: 23).")
```
