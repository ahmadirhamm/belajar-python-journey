# Modul 11: Visualisasi Data dengan Matplotlib dan Seaborn

Visualisasi data digunakan untuk menyajikan data dalam bentuk grafis. Kita akan menggunakan dua library utama:
1. **Matplotlib**: Library dasar untuk membuat grafik (sangat fleksibel dan kuat).
2. **Seaborn**: Library tingkat lanjut yang dibangun di atas Matplotlib, dirancang untuk membuat grafik statistik yang jauh lebih indah dan modern secara otomatis.

````python
# 1. IMPORT LIBRARY VISUALISASI
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Kita gunakan tema bawaan Seaborn agar grafik otomatis terlihat modern
sns.set_theme(style="darkgrid")

# Data simulasi perkembangan jaringan (misal: jumlah user terkoneksi perbulan)
bulan = ["Jan", "Feb", "Mar", "Apr", "Mei", "Jun"]
user_aktif = [120, 150, 180, 240, 310, 420]

print("--- Membuat Line Chart (Grafik Garis) ---")
# Menentukan ukuran kanvas grafik (lebar, tinggi)
plt.figure(figsize=(8, 4))

# Membuat grafik garis
plt.plot(bulan, user_aktif, marker='o', color='purple', linewidth=2)

# Menambah judul dan label sumbu
plt.title("Tren Pertumbuhan Aktif Jaringan (2026)", fontsize=14, fontweight='bold')
plt.xlabel("Bulan", fontsize=12)
plt.ylabel("Jumlah User", fontsize=12)

# Menampilkan Grafik
plt.show()
````

## 2. Membuat Bar Chart (Grafik Batang) dengan Seaborn
Bar chart sangat cocok digunakan untuk membandingkan nilai antar kategori data yang berbeda.

````python
# Data simulasi jenis perangkat di ruang server
data_perangkat = {
    "Kategori" : ["Switch", "Router", "Server", "Access Point"],
    "Jumlah Unit" : [15, 8, 5, 22]
}
df_server = pd.DataFrame(data_perangkat)

print("--- Membuat Bar Chart dengan Seaborn ---")
plt.figure(figsize=(8, 4))

# Membuat Grafik Batang menggunakan Seaborn
sns.barplot(x="Kategori", y="Jumlah Unit", data=df_server, palette="magma")

# Menambahkan detail teks
plt.title("Distribusi Perangkat Infrastruktur Jaringan", fontsize=14, fontweight='bold')
plt.xlabel("Jenis Perangkat", fontsize=12)
plt.ylabel("Total Unit", fontsize=12)

plt.show()
````

## 📝 Tantangan Modul 11 (Grafik Penjualan Produk)
Mari kita visualisasikan data performa penjualan menu di sebuah coffee shop:
1. Buatlah DataFrame dari dictionary berikut:
   ```python
   data_kopi = {
       "Menu": ["Espresso", "Americano", "Caffe Latte", "Cappuccino", "Matcha Latte"],
       "Terjual (Cup)": [40, 65, 120, 85, 55]
   }

2. Buatlah sebuah Bar Chart (Grafik Batang) menggunakan Seaborn (sns.barplot).

3. Set sumbu x untuk Menu dan sumbu y untuk Terjual (Cup).

4. Berikan judul grafik bebas, contoh: "Performa Penjualan Menu Coffee Shop".

5. Gunakan palet warna yang menarik (pilihan palet bebas, misal: "viridis", "rocket", atau "crest").

````python
# Jawaban modul 11 (Grafik Penjualan Produk)
data_kopi = {
    "Menu": ["Espresso", "Americano", "Caffe Latte", "Cappuccino", "Matcha Latte"],
    "Terjual (Cup)": [40, 65, 120, 85, 55]
}
df_kopi = pd.DataFrame(data_kopi)

print("--- Performa Penjualan Menu Coffee Shop ---")
plt.figure(figsize=(8, 5))

# Membuat Grafik Batang menggunakan Seaborn (Ditambahkan hue dan legend=False untuk fix FutureWarning)
sns.barplot(x="Menu", y="Terjual (Cup)", data=df_kopi, palette="rocket", hue="Menu", legend=False)

# Menambahkan Detail Teks Informasi
plt.title("Performa Penjualan Menu Coffee Shop", fontsize=14, fontweight='bold')
plt.xlabel("Menu Minuman", fontsize=12)
plt.ylabel("total Terjual (Cup)", fontsize=12)

# Menampilkan Grafik
plt.show()
````
