# Modul 6: Fungsi (Functions - def)

Fungsi adalah blok kode terorganisir yang digunakan untuk melakukan satu tindakan spesifik.
Keuntungan menggunakan fungsi:
1. **Reusability**: Kode cukup ditulis sekali, tapi bisa digunakan berkali-kali.
2. **Clean Code**: Membuat struktur kode kita jauh lebih rapi dan mudah dibaca.

Di Python, fungsi didefinisikan menggunakan kata kunci `def`.

```python
# 1. FUNGSI TANPA PARAMETER (Fungsi Paling Sederhana)
def sapa_user():
  print("Hallo Ahmad!")
  print("Selamat datang di modul 6 Python.")

# Cara menggunakannya (memanggil fungsi):
print("--- Memanggil Fungsi Sapa ---")
sapa_user()
```

## 2. Fungsi dengan Parameter/Argumen
Kita bisa memasukkan variabel (input) ke dalam fungsi agar hasil kodenya bisa berubah-ubah sesuai input yang kita berikan.

```python
# 2. FUNGSI WITH PARAMETER
def sapa_peserta(nama, modul_ke):
  print(f"Hallo {nama}, semangat belajar Modul {modul_ke}-nya ya!")

print("--- Memanggil Fungsi dengan Input berbeda ---")
sapa_peserta("Ahmad", 6)
sapa_peserta("Irham", 3)
```

## 3. Fungsi dengan Pengembalian Nilai (`return`)
Ini adalah jenis fungsi yang paling sering digunakan dalam industri. Kata kunci `return` digunakan agar fungsi tidak sekadar mencetak teks ke layar, melainkan **menghasilkan suatu nilai** yang bisa kita simpan ke dalam variabel lain untuk diolah lagi nanti.

```python
# 3. FUNGSI DENGAN RETURN
def hitung_diskon(harga_total):
  diskon = harga_total * 0.1 # Diskon 10%
  return diskon

# memanggil fungsi dan menyimpan hasilnya ke variabel baru
total_belanja = 200000
potongan_harga = hitung_diskon(total_belanja)

print("--- Perhitungan Kasir Bersama Fungsi ---")
print(f"Total Belanja : Rp {total_belanja}")
print(f"Diskon (10%)  : Rp {potongan_harga}")
print(f"Harus Dibayar : Rp {total_belanja - potongan_harga}")
```

## 📝 Tantangan Modul 6 (Konverter Suhu)
Mari kita buat alat konversi suhu otomatis dari Celsius ke Fahrenheit menggunakan fungsi:
1. Buatlah sebuah fungsi bernama `celsius_ke_fahrenheit` yang menerima satu parameter bernama `celsius`.
2. Di dalam fungsi tersebut, hitung konversinya menggunakan rumus: `(celsius * 9/5) + 32`
3. Gunakan kata kunci **`return`** untuk mengembalikan hasil perhitungan tersebut.
4. Di luar fungsi, buat variabel `suhu_celsius = 30`.
5. Panggil fungsi Anda dengan memasukkan variabel `suhu_celsius` tersebut, lalu simpan hasilnya ke variabel `hasil_fahrenheit`.
6. Cetak menggunakan f-string: *"... derajat Celsius sama dengan ... derajat Fahrenheit."*

```python
# Jawaban tantangan modul 6 (konverter suhu)
def celsius_ke_fahrenheit(celsius):
  fahrenheit = (celsius * 9 / 5)  + 32
  return fahrenheit

# # memanggil fungsi
suhu_celsius = 30
hasil_fahrenheit = celsius_ke_fahrenheit(suhu_celsius)

print(f"{suhu_celsius} derajat celsius sama dengan {hasil_fahrenheit} derajat fahrenheit.")
```

Catatan Tambahan: Python otomatis mengubah hasil pembagian / menjadi tipe data desimal (float), itulah mengapa hasil Anda keluar 86.0 (ada angka di belakang koma). Penulisan di atas membuatnya jauh lebih rapi saat dibaca oleh programmer lain!
