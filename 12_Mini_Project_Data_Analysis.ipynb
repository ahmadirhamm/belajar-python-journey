Proyek ini bertujuan untuk menganalisis data performa perangkat jaringan di 5 area kerja kantor cabang, menyaring area yang mengalami gangguan, dan memvisualisasikannya agar tim infrastruktur IT bisa melakukan optimasi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set tema grafik
sns.set_theme(style="darkgrid")

# Data mentah log jaringan kantor cabang
log_jaringan = {
    "Area Kerja": ["Ruang Utama", "Ruang Server", "Lantai 2", "Gedung A", "Gedung B"],
    "Bandwidth_Mbps": [150, 450, 100, 250, 80],
    "Rata_Rata_Ping_ms": [15, 8, 85, 22, 120], # Ping di atas 80ms mengindikasikan lag/gangguan
    "Status_Link": ["Aktif", "Aktif", "Gangguan", "Aktif", "Gangguan"]
}

# Mengubah menjadi DataFrame
df_network = pd.DataFrame(log_jaringan)
df_network
```

📝 INSTRUKSI TUGAS (Wajib Dikerjakan Menggunakan Kode)
Buatlah Code Cell baru di bawahnya untuk menyelesaikan 3 Langkah Analisis berikut:

🛠️ Langkah 1: Analisis Statistik (NumPy & Pandas Essentials)
- Hitung dan cetak Rata-rata Bandwidth dari seluruh area kerja menggunakan fungsi statistik (.mean()).

- Hitung dan cetak Total Bandwidth yang dialokasikan di kantor tersebut menggunakan fungsi .sum().

🔍 Langkah 2: Pembersihan & Penyaringan Data (Filtering)
- Saring (filter) DataFrame df_network untuk menampilkan area kerja yang Status_Link-nya bertuliskan "Gangguan" saja. Simpan hasilnya ke dalam variabel bernama df_trouble.

- Tampilkan tabel df_trouble tersebut ke layar.

📊 Langkah 3: Visualisasi Data (Matplotlib & Seaborn)
- Buatlah sebuah Bar Chart (Grafik Batang) menggunakan sns.barplot.

- Sumbu X = "Area Kerja", Sumbu Y = "Rata_Rata_Ping_ms", data diambil dari DataFrame utama (df_network).

- Gunakan parameter palette="coolwarm", serta tambahkan hue="Area Kerja" dan legend=False agar kode bersih dari FutureWarning.

- Berikan Judul Grafik: "Analisis Latency (Ping) per Area Kerja".

- Berikan Label Sumbu X: "Lokasi Area" dan Label Sumbu Y: "Ping (ms)".

```python
# =====================================================================
# JAWABAN MINI PROJECT DATA ANALYSIS (FINAL EXAM)
# =====================================================================

# --- LANGKAH 1: ANALISIS STATISTIK ---

# Menghitung rata-rata Bandwidth dari seluruh area kerja menggunakan .mean()
print("--- Tampilan rata-rata Bandwidth dan total bandwidth yang terpakai ---")

rata_rata = df_network["Bandwidth_Mbps"].mean()
print(f"Rata-rata Bandwidth dari seluruh area kerja adalah: {rata_rata} Mbps.")

# Menghitung total Bandwidth yang di alokasikan di kantor tersebut menggunakan fungsi .sum()
Total_Bandwidth = df_network["Bandwidth_Mbps"].sum()
print(f"Total Bandwidth yang di alokasikan di kantor tersebut adalah: {Total_Bandwidth} Mbps.")
```

```python
# --- LANGKAH 2: FILTERING DATA ---

# Filter DataFrame df_network untuk menampilkan area kerja yang berstatus "Gangguan".
# Tampilan tabel status link yang gangguan.

df_trouble = df_network[df_network["Status_Link"] == "Gangguan"]

# Menampilkan baris yang gangguan saja
print("--- Menampilkan Tabel hasil filter area yang bermasalah")
print(df_trouble)
print("\n")
```

```python
# --- LANGKAH 3: VISUALISASI DATA (Matplotlib & Seaborn) ---
print("--- Membuat Line Chart (Grafik Garis) ---")
# Menentukan ukuran kanvas grafik (lebar, tinggi)
plt.figure(figsize=(8, 4))

# Membuat grafik garis
sns.barplot(
    x="Area Kerja",
    y="Rata_Rata_Ping_ms",
    data=df_network,
    palette="coolwarm",
    hue="Area Kerja",
    legend=False
)

# Menambah judul dan label sumbu
plt.title("Analisis Latency (Ping) per Area Kerja", fontsize=14, fontweight='bold')
plt.xlabel("Lokasi Area", fontsize=12)
plt.ylabel("Ping (ms)", fontsize=12)

# Menampilkan Grafik
plt.show()
```

```python
# =====================================================================
# JAWABAN MINI PROJECT DATA ANALYSIS (FINAL EXAM)
# =====================================================================

# --- LANGKAH 1: ANALISIS STATISTIK ---
rata_rata_bw = df_network["Bandwidth_Mbps"].mean()
total_bw = df_network["Bandwidth_Mbps"].sum()

print("--- Langkah 1: Analisis Statistik ---")
print(f"Rata-rata Bandwidth : {rata_rata_bw} Mbps")
print(f"Total Alokasi Bandwidth : {total_bw} Mbps\n")


# --- LANGKAH 2: FILTERING DATA ---
df_trouble = df_network[df_network["Status_Link"] == "Gangguan"]

print("--- Langkah 2: Menyaring Perangkat Gangguan ---")
# Menampilkan tabel hasil filter area yang bermasalah
print(df_trouble)
print("\n")


# --- LANGKAH 3: VISUALISASI DATA ---
print("--- Langkah 3: Membuat Grafik Batang Latency ---")
plt.figure(figsize=(8, 5))

# Membuat bar chart dengan Seaborn sesuai instruksi
sns.barplot(
    x="Area Kerja",
    y="Rata_Rata_Ping_ms",
    data=df_network,
    palette="coolwarm",
    hue="Area Kerja",
    legend=False
)

# Menambahkan elemen teks informasi grafik
plt.title("Analisis Latency (Ping) per Area Kerja", fontsize=14, fontweight="bold")
plt.xlabel("Lokasi Area", fontsize=12)
plt.ylabel("Ping (ms)", fontsize=12)

# Menampilkan grafik ke layar
plt.show()
```
