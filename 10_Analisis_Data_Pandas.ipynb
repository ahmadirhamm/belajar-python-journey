# Modul 10: Analisis Data dengan Pandas

Pandas adalah library Python yang paling populer untuk analisis data dan data wrangling.

Dua struktur data utama pada Pandas:
1. **Series**: Array 1 dimensi (seperti satu kolom tunggal di Excel).
2. **DataFrame**: Tabel 2 dimensi (memiliki baris dan kolom, sangat mirip dengan lembar kerja Excel).

````python
# 1. IMPORT PANDAS & MEMBUAT DATAFRAME MANUAL
# 'as pd' adalah alias standar industri untuk library Pandas
import pandas as pd

# Membuat Dataframe dari Dictionary
data_toko = {
    "Nama Produk": ["Kopi susu", "Roti Bakar", "Kentang Goreng"],
    "Harga": [15000, 18000, 12000],
    "Stok": [50, 20, 35]
}

df = pd.DataFrame(data_toko)

print("--- DataFrame Pandas Pertama Anda ---")
# Di Colab, cukup ketik 'df' di baris terakhir untuk menampilkan tabel yang estetik
df
````

## 2. Mengambil Data Spesifik (Filtering & Selection)
Kita bisa memilih kolom tertentu atau menyaring baris berdasarkan kondisi logika yang kita inginkan dengan sangat mudah.

````python
print("--- Mengambil Satu Kolom (series) ---")
print(df["Nama Produk"])

print("\n--- Menyaring Data (Harga > 13.000) ---")
# Menampilkan baris yang harga nya di atas 13000 saja
produk_mahal = df[df["Harga"] > 13000]
produk_mahal
````

## 3. Analisis Deskriptif Cepat
Pandas memiliki fungsi ajaib bernama `.describe()` yang bisa langsung mengeluarkan ringkasan statistik dari semua kolom yang berisi angka secara otomatis.

````python
print("--- Ringkasan Statistik DataFrame ---")
df.describe()
````
