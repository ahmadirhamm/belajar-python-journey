# Modul 7: Object-Oriented Programming (OOP) Dasar

Pemrograman Berbasis Objek (OOP) adalah cara mengatur kode dengan membungkus data (Atribut) dan fungsi (Method) ke dalam satu kesatuan yang disebut **Objek**.

Konsep utama OOP Dasar:
1. **Class**: Cetakan atau cetak biru (blueprint) untuk membuat objek.
2. **Object**: Hasil nyata yang dibuat dari cetakan (Class) tersebut.
3. **Attribute**: Variabel/data yang menempel pada objek (misal: nama, warna, spek).
4. **Method**: Fungsi/aksi yang bisa dilakukan oleh objek tersebut.

```python
# 1. MEMBUAT CLASS DAN OBJECT SEDERHANA
class RobotSederhana:
  pass # 'pass' digunakan sementara agar tidak error saat class masih kosong

# Membuat object dari Class RobotSederhana
robot_ahmad = RobotSederhana()
print("Tipe dari robot_ahmad:", type(robot_ahmad))
```

## 2. Menggunakan Constructor `__init__` dan Atribut
Di Python, kita menggunakan fungsi khusus bernama `__init__` (disebut constructor) untuk mengisi data/atribut awal saat objek pertama kali diciptakan. Kata kunci `self` wajib digunakan untuk menandakan bahwa atribut itu milik objek tersebut.

```python
# 2. CLASS DENGAN DATA (ATRIBUT) & AKSI (METHOD)
class Hero:
  # Constructor untuk menginisialisasi atribut
  def __init__(self, nama, peran, darah):
    self.nama = nama   # Atribut Nama
    self.peran = peran # Atribut Peran/Role
    self.darah = darah # Atribut Darah/HP

  # Method (aksi yang bisa di lakukan hero)
  def perkenalan(self):
    print(f"Saya {self.nama}, seorang {self.peran}, dengan HP {self.darah}!")

# Membuat 2 objek hero berada dari 1 cetakan yang sama
hero_1 = Hero("Layla", "Marskman", 2500)
hero_2 = Hero("Tigreal", "Tank", 4000)

print("--- Aksi Para Hero ---")
hero_1.perkenalan()
hero_2.perkenalan()
```

## 📝 Tantangan Modul 7 (Sistem Akun User)
Mari kita simulasikan pembuatan sistem objek untuk akun user di sebuah platform aplikasi:
1. Buatlah sebuah class bernama `User`.
2. Di dalam class tersebut, buat constructor `__init__` yang menerima atribut: `username`, `email`, dan `is_admin` (boolean).
3. Buat sebuah method bernama `tampilkan_profil` di dalam class tersebut yang akan mencetak kalimat menggunakan f-string:
   *"User {username} ({email}) memiliki status admin: {is_admin}"*
4. Di luar class, buatlah objek baru bernama `user_baru` dengan data bebas (misal: username "irham11", email "irham@email.com", is_admin True).
5. Panggil method `tampilkan_profil` dari objek `user_baru` tersebut.

```python
# Jawaban tantangan modul 7
class User:
  # Constructor untuk menginisialisasi atribut
  def __init__(self, username, email, is_admin):
    self.username = username   # Atribut username
    self.email = email         # Atribut email
    self.is_admin = is_admin   # Atribut admin

  # Aksi Method yang akan di tampilkan
  def tampilkan_profil(self):
    print(f"User {self.username}, dengan nama email ({self.email}), memiliki status admin: {self.is_admin}")


# Objek baru
user_baru = User("irham11", "irham11@gmail.com", is_admin=True)

# Menampilkan Akun user baru dan memanggil method tampilkan_profil
print("--- Akun User Baru ---")
user_baru.tampilkan_profil()
```
