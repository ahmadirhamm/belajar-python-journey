# Modul 4: Percabangan (If, Elif, Else)

Percabangan digunakan untuk membuat program mengambil keputusan berdasarkan kondisi tertentu.
Python akan memeriksa apakah suatu kondisi bernilai `True` (Benar) atau `False` (Salah), lalu menjalankan kode yang sesuai.

*Penting:* Python sangat sensitif dengan **Indentasi** (jarak spasi/tab di awal baris). Kode di dalam blok `if` harus menjorok ke dalam.

```python
# 1. Dasar IF-ELSE
Nilai_ujian =78

print("--- Hasil Kelulusan ---")
if Nilai_ujian >= 75:
  # kode ini hanya jalan jika nilai ujian di atas 75
  print("Selamat! Anda dinyatakan LULUS.")
else :
  # kode ini jalan jika kondisi di atas TIDAK terpenuhi
  print("Maaf, Anda harus mengulang ujian.")
```

## 2. Menggunakan Banyak Kondisi (Elif)
Jika pilihan kondisinya lebih dari dua, kita bisa menggunakan `elif` (short untuk else if).

```python
# 2. CONTOH ELIF  (Konversi Nilai ke huruf)
skor = 89

print("--- Evaluasi Nilai ---")
if skor >= 90:
  grade = "A"
elif skor >= 80:
  grade = "B"
elif skor >= 70:
  grade = "C"
else:
  grade = "D"

print(f"Skor Anda: {skor}, Grade Anda: {grade}")
```

## 3. Kombinasi Operator Logika (`and`, `or`)
Kita bisa menggabungkan dua atau lebih kondisi sekaligus.
- `and`: Semua kondisi HARUS benar agar kodenya jalan.
- `or`: Minimal salah satu kondisi benar, kodenya sudah bisa jalan.

```python
# Contoh penggunaan operator LOGIKA
umur = 20
punya_sim = True

print("--- Pemeriksaan Polisi ---")
if umur >= 17 and punya_sim == True:
  print("Anda diizinkan berkendara.")
else:
  print("Anda TIDAK boleh berkendara.")
```

## 📝 Tantangan Modul 4 (Sistem Akun Premium)
Buatlah sebuah sistem logika sederhana untuk menentukan apakah seorang user bisa menonton video kualitas 4K:
1. Buat variabel `is_premium` (isi dengan `True` atau `False`).
2. Buat variabel `kecepatan_internet` (isi dengan angka, misal `30` dalam Mbps).
3. Buatlah kondisi `if-else` menggunakan operator **`and`**:
   - Jika `is_premium` bernilai `True` **DAN** `kecepatan_internet` lebih besar atau sama dengan (`>=`) `25`, cetak kalimat: *"Streaming 4K berhasil dimulai!"*
   - Jika tidak memenuhi syarat, cetak kalimat: *"Gagal streaming 4K. Pastikan akun Anda Premium dan internet minimal 25 Mbps."*

```python
# Jawaban tantangan modul 4
kecepatan_internet = 30 # (=) operator penugasan
is_premium = True

print("--- Streaming Video ---")
if kecepatan_internet >= 25 and is_premium == True: # (==) operator perbandingan True/False
  print("Streaming 4K berhasil dimulai!")
else:
  print("Gagal streaming 4K. Pastikan akun anda premium dan internet minimal 25 Mbps.")

```

Tips Pro Tambahan untuk Kode Anda:
Di Python, ada cara penulisan yang jauh lebih elegan (Pythonic way) khusus untuk variabel Boolean (True/False).

Karena variabel is_premium itu sendiri sudah bernilai True, Anda tidak perlu lagi menulis == True. Anda bisa langsung menulisnya seperti ini:

# Cara penulisan yang lebih bersih dan profesional:
if kecepatan_internet >= 25 and is_premium:
    print("Streaming 4K berhasil dimulai!")
