## A. Variabel
Variabel adalah wadah untuk menyimpan nilai data. Di Python, Anda tidak perlu mendeklarasikan tipe variabel secara eksplisit seperti pada bahasa C++ atau Java. Cukup tulis nama variabel dan gunakan tanda `=` untuk mengisinya.

```python
# Membuat Variabel
nama = "Ahmad"
umur = 23
tinggi_badan = 169.5
is_student = True

# Menampilkan isi variabel
print("Nama:", nama)
print("Umur:", umur, "tahun")
print("Tinggi:", tinggi_badan, "cm")
print("Status Mahasiswa:", is_student)
```

## B. Tipe Data Dasar
Python memiliki beberapa tipe data bawaan yang sering digunakan:
1. `str` (String): Teks/karakter, diapit tanda kutip.
2. `int` (Integer): Bilangan bulat.
3. `float`: Bilangan pecahan/desimal (menggunakan titik, bukan koma).
4. `bool` (Boolean): Bernilai `True` atau `False`.

Kita bisa mengecek tipe data suatu variabel menggunakan fungsi `type()`.

```python
# Mengecek tipe data dari variabel yang telah dibuat sebelumnya
print(type(nama))          # Output: <class 'str'>
print(type(umur))          # Output: <class 'int'>
print(type(tinggi_badan))  # Output: <class 'float'>
print(type(is_student))    # Output: <class 'bool'>
```

## C. Tantangan Mandiri (Mini Challenge)
Cobalah buat 3 variabel baru yang mendeskripsikan sebuah komputer/laptop:
1. `brand` (String)
2. `ram` (Integer)
3. `is_ssd` (Boolean)

Kemudian cetak variabel tersebut ke layar beserta tipe datanya menggunakan fungsi `print()` dan `type()`.

```python
# Jawaban mini challenge
brand = "Acer"
ram = 16
is_ssd = True

# Menampilkan Isi variabel
print("Brand:", brand)
print("Ram:", ram, "GB")
print("Ssd:", is_ssd)
```

```python
# Mengecek tipe data dari variabel mini challenge di atas
print(type(brand))   # Output : <class 'str'>
print(type(ram))     # Output : <class 'int'>
print(type(is_ssd))  # Output : <class 'bool'>
```
