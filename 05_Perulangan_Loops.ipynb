# Modul 5: Perulangan (For dan While Loops)

Perulangan (looping) digunakan untuk mengeksekusi blok kode secara berulang selama kondisi tertentu terpenuhi.
Di Python, ada dua jenis perulangan utama:
1. **For Loop**: Digunakan jika kita sudah tahu pasti berapa kali kode harus diulang (misal: mengulang sebanyak isi List).
2. **While Loop**: Digunakan jika jumlah perulangannya belum pasti, dan bergantung pada suatu kondisi logic (akan terus mengulang *selama* kondisi bernilai True).

```python
# 1. FOR LOOP (Menggunakan fungsi range() atau elemen list)
print("--- For Loop dengan range() ---")
# range(5) artinya mengulang dari angka 0 sampai angka 4 (total 5 kali)
for i in range(5):
  print(f"Perulangan ke-{i}")

print("\n--- For Loop menelusuri isi list ---")
peserta = ["Ahmad", "Irham", "Dzulkifli"]
for nama in peserta:
  print(f"Selamat Belajar, {nama}!")
```

## 2. While Loop
While loop akan terus berjalan selama kondisinya masih `True`. **Penting:** Pastikan ada kode yang mengubah kondisi tersebut di dalam blok perulangan agar tidak terjadi *infinite loop* (perulangan tanpa henti yang bisa membuat browser/komputer hang).

```python
# 2. WHILE LOOP
hitung = 1

print("--- Berhitung dengan While Loop ---")
while hitung <= 5:
  print(f"Angka: {hitung}")
  hitung = hitung + 1 # Mengisi 'pancingan' agar kondisi beralih ke False saat mencapai angka 6
```

## 3. Kontrol Perulangan (`break` dan `continue`)
- `break`: Digunakan untuk menghentikan paksa perulangan saat itu juga.
- `continue`: Digunakan untuk melompati sisa kode di perulangan saat ini dan langsung loncat ke perulangan berikutnya.

```python
print("--- Contoh Break ---")
for angka in range(1, 10):
  if angka == 5:
    break # Perulangan langsung BERHENTI total ketika menyentuh angka 5
  print(angka)

print("\n--- Contoh Continue ---")
for angka in range(1, 6):
  if angka == 3:
    continue # Angka 3 akan DILOMPATI/DIABAIKAN, langsung lanjut ke angka 4
  print(angka)
```

## 📝 Tantangan Modul 5 (Aplikasi Kasir Sederhana)
Bayangkan Anda diminta membuat sistem kasir yang otomatis menghitung total belanjaan dari sebuah keranjang:
1. Buat sebuah list bernama `harga_barang` berisi angka: `[15000, 25000, 50000, 10000]`
2. Buat variabel `total_bayar` dengan nilai awal `0`.
3. Gunakan **For Loop** untuk menjumlahkan semua angka yang ada di dalam list `harga_barang` ke dalam variabel `total_bayar`.
4. Cetak hasil akhirnya menggunakan f-string: *"Total belanjaan Anda hari ini adalah Rp {total_bayar}"*

*Petunjuk Tambahan:* Di dalam loop, Anda bisa menggunakan perintah `total_bayar = total_bayar + harga` (atau disingkat `total_bayar += harga`) untuk menjumlahkannya secara akumulatif.

```python
# Jawaban tantangan modul 5

harga_barang = ["15000", "25000", "50000", "10000"] # contoh penulisan list yang salah karena list tersebut berisikan angka jadi tidak perlu menggunakan tanda ("").
total_bayar = 0

for r in harga_barang: # Huruf r adalah sebuah nama variabel sementara (sering disebut iterator atau loop variable).
  total_bayar = total_bayar + int(r) # menggunakan fungsi int(r) untuk mengubahnya kembali menjadi angka bulat sebelum dijumlahkan.

print(f"Total belanjaan Anda hari ini adalah Rp {total_bayar}.")
```

Tips "Expert Mindset" untuk Penamaan Variabel Loop:
Meskipun menggunakan huruf tunggal seperti r atau i itu sah dan kodenya berjalan lancar, kode Anda akan jauh lebih mudah dibaca jika menggunakan nama yang deskriptif (menggambarkan isi datanya).

Aturan praktisnya: Jika nama list-nya berbentuk jamak (contoh: harga_barang), maka variabel di dalam loop-nya sebaiknya berbentuk tunggal (contoh: harga).

Perhatikan perbandingan ini:

```python
# Menggunakan 'r' (Komputer paham, tapi manusia harus mikir 'r' itu apa)
for r in harga_barang:
    total_bayar = total_bayar + int(r)

# Menggunakan 'harga' (Sangat bersih dan mudah dipahami siapa saja)
for harga in harga_barang:
    total_bayar = total_bayar + int(harga)
```
