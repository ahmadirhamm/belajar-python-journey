# Modul 3: Struktur Data - List dan Dictionary

Jika variabel biasa hanya bisa menyimpan satu nilai, Struktur Data memungkinkan kita menyimpan sekumpulan data.
Dua yang paling sering digunakan adalah:
1. **List**: Urutan data yang bisa diubah (Mutable).
2. **Dictionary**: Data berbasis pasangan kunci dan nilai (Key-Value).

```python
# 1. LIST (menggunakan kurung siku [])
buah = ["apel", "jeruk", "mangga", "pisang"]

print("--- Belajar List ---")
print("daftar buah:", buah)
print("buah pertama (index 0):", buah[0]) # di python hitungan di mulai dari 0
print("buah terakhir         :", buah[-1])

# menambah dan menghapus data list
buah.append("semangka") # menambah di akhir
buah[1] = "alpukat"     # mengubah isi index ke-1
del buah[2]             # menghapus mangga

print("list setelah di update:", buah)
```

## 2. DICTIONARY
Dictionary sangat berguna untuk menyimpan data yang memiliki label, seperti profil user atau spesifikasi barang. Menggunakan kurung kurawal `{ }`.

```python
# 2. Dictionary (key: value)
user = {
    "nama": "Ahmad",
    "level": "Beginner",
    "is_active": "True",
    "skor": "85"
}

print("--- Belajar Dictionary ---")
print("Nama User :", user["nama"])
print("Level     :", user["level"])

# menambah key baru
user["bahasa"] = "Python"

# mengupdate nilai
user["level"] = "Intermediated"

print("Data User Terbaru:", user)
```

## 📝 Tantangan Modul 3
Bayangkan Anda sedang membuat sistem database untuk sebuah smartphone:
1. Buat variabel `smartphone` dalam bentuk **Dictionary** yang berisi:
   - `brand` (string)
   - `storage` (integer, misal 128)
   - `warna` (list berisi 3 warna, misal: "Hitam", "Putih", "Biru")
2. Ubah `storage` menjadi 256.
3. Cetak kalimat menggunakan f-string:
   "Smartphone {brand} memiliki pilihan warna {warna pertama dari list} dan kapasitas {storage}GB."

```python
# Jawaban tantangan 3
smartphone = {
    "brand": "Iphone",
    "storage": 128,
    "warna": ["Hitam", "putih", "oranye"]
}

#mengupddate storage
smartphone ["storage"] = 256

# memanggil data dengan f-string
# penulisan pemanggil variabel dictionary: menggunakan ' supaya tidak bentrok dengan " f-string
print(f"Smartphone {smartphone['brand']} memiliki pilihan warna {smartphone['warna'][0]} dan kapasitas {smartphone['storage']} GB.")
```

CATATAN:

smartphone['brand']: Mengambil nilai "Iphone".

smartphone['warna'][0]: Ini artinya: "Buka dictionary smartphone, cari kunci warna, lalu ambil item pertama ([0]) dari daftar warna tersebut."

Kutip Satu vs Kutip Dua: Di dalam f"..." (kutip dua), saat memanggil kunci dictionary harus pakai kutip satu '...' supaya Python tidak bingung di mana teks itu berakhir.
